```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
y_naive=model.predict(x_test)
y_naive
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
mat=confusion_matrix(y_test,y_naive)
mat
Naivescore=accuracy_score(y_test,y_naive)
Naivescore
report=classification_report(y_test,y_naive)
report
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=6)
model.fit(x_train,y_train)
y_neighbors=model.predict(x_test)
y_neighbors
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
mat=confusion_matrix(y_test,y_neighbors)
mat
Neighbors_score=accuracy_score(y_test,y_neighbors)
Neighbors_score
report=classification_report(y_test,y_neighbors)
report
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_svm=model.predict(x_test)
y_svm
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
mat=confusion_matrix(y_test,y_svm)
mat
SVC_score=accuracy_score(y_test,y_svm)
SVC_score
report=classification_report(y_test,y_svm)
report
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
y_tree=model.predict(x_test)
y_tree
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
mat=confusion_matrix(y_test,y_tree)
mat
Tree_score=accuracy_score(y_test,y_tree)
Tree_score
report=classification_report(y_test,y_tree)
report
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
y_ran=model.predict(x_test)
y_ran
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
mat=confusion_matrix(y_test,y_ran)
mat
Random_score=accuracy_score(y_test,y_ran)
Random_score
report=classification_report(y_test,y_ran)
report
```

```python
alg=['KNN','Naive_bayes','SVM','DecisionTree','RandomForest']
acc=[Neighbors_score,Naivescore,SVC_score,Tree_score,Random_score]
accuracy_score=pd.DataFrame({'Algorithms':alg,'Accuracy':acc})
accuracy_score['Accuracy']=accuracy_score['Accuracy']*100
accuracy_score
```

```python
accuracy_score.sort_values(by='Accuracy',ascending=False)
```

```python
ax=sns.barplot(y='Algorithms',x='Accuracy',palette='muted',data=accuracy_score.sort_values(by='Accuracy',ascending=False)
)
for i in ax.containers:
  ax.bar_label(i,)
  plt.tight_layout()
```
